## 查看metadata的（y）性能提升

```python
improve_accuracy_max = np.max(process_australian_metadata[:, 396])
improve_accuracy_min = np.min(process_australian_metadata[:, 396])
print('improve_accuracy_max:  ', improve_accuracy_max)
print('improve_accuracy_min:  ', improve_accuracy_min)

improve_accuracy_mean = np.mean(process_australian_metadata[:, 396])
improve_accuracy_std = np.std(process_australian_metadata[:, 396])
print('improve_accuracy_mean:  ', improve_accuracy_mean)
print('improve_accuracy_std:  ', improve_accuracy_std)

postive_improve_accuracy_mean = np.mean(process_australian_metadata[np.where(process_australian_metadata[:, 396]>0)[0], 396])
postive_improve_accuracy_std = np.std(process_australian_metadata[np.where(process_australian_metadata[:, 396]>0)[0], 396])
print('postive_improve_accuracy_mean:  ', postive_improve_accuracy_mean)
print('postive_improve_accuracy_std:  ', postive_improve_accuracy_std)

negative_improve_accuracy_mean = np.mean(process_australian_metadata[np.where(process_australian_metadata[:, 396]<0)[0], 396])
negative_improve_accuracy_std = np.std(process_australian_metadata[np.where(process_australian_metadata[:, 396]<0)[0], 396])
print('negative_improve_accuracy_mean:  ', negative_improve_accuracy_mean)
print('negative_improve_accuracy_std:  ', negative_improve_accuracy_std)
```
