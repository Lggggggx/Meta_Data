# 分析设计的metadata信息与模型性能提升之间的关系

根据生成的metadata信息，用一个RandomForestRegressor去拟合metadata与模型performance improvement之间的对应关系

利用随机森林回归器的特性，能够显示各个特征的重要程度

```python
import numpy as np

import matplotlib.pyplot as plt

from sklearn.externals import joblib

rfc_p_classify_australian = joblib.load('./rfc_p_classify_australian.joblib')
rfc_p_classify_ethn = joblib.load('./rfc_p_classify_ethn.joblib')
rfr_p_regression_australian = joblib.load('./rfr_p_regression_australian.joblib')
rfr_p_regression_ethn = joblib.load('./rfr_p_regression_ethn.joblib')
rfr_p_regression_wdbc = joblib.load('./rfr_p_regression_wdbc.joblib')
#  = jolib.load('./.joblib')
#  = jolib.load('./.joblib')
#  = jolib.load('./.joblib')
#  = jolib.load('./.joblib')
# rfr = joblib.load('./rfr_p_regression_wdbc.joblib')
```

```python
x = np.arange(0,396)

# plt.subplot(231)
plt.scatter(x, rfc_p_classify_australian.feature_importances_)
plt.title('rfc_p_classify_australian')

# plt.subplot(232)
plt.scatter(x, rfc_p_classify_ethn.feature_importances_)
plt.title('rfc_p_classify_ethn')

# plt.subplot(234)
plt.scatter(x, rfr_p_regression_australian.feature_importances_)
plt.title('rfr_p_regression_australian')

# plt.subplot(235)
plt.scatter(x, rfr_p_regression_ethn.feature_importances_)
plt.title('rfr_p_regression_ethn')

# plt.subplot(236)
plt.scatter(x, rfr_p_regression_wdbc.feature_importances_)
plt.title('rfr_p_regression_wdbc')

plt.figure(figsize=(100, 100))
plt.show()
```
